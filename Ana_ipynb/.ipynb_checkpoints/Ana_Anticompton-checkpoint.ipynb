```c++
TFile *ipf = new TFile("../root/g4anticompton.root");
TTree *tree = (TTree*)ipf->Get("tree");
TCanvas *c1 = new TCanvas;
```

```c++
tree->Scan("ghit:gEventID:ge0:ge:gt:gid","","colsize=8",10,0);
```

```c++
tree->Scan("ahit:aEventID:ae0:ae:at:aid","","colsize=8",1,2188);
```

```c++
%jsroot on
tree->Draw("ae>>ha(2000,0,2000)");
tree->Draw("ae>>hab(2000,0,2000)","abs(ae-ae0)>=0.01");
TH1D *ha = (TH1D*)gROOT->FindObject("ha");
TH1D *hab = (TH1D*)gROOT->FindObject("hab");
ha->Draw();
hab->SetLineColor(kRed);
hab->Draw("same");
c1->Draw();
```

```c++
%jsroot on
tree->Draw("ge>>hg(2000,0,2000)");
tree->Draw("ge>>hgb(2000,0,2000)","abs(ge-ge0)>=0.01");
TH1D *hg = (TH1D*)gROOT->FindObject("hg");
TH1D *hgb = (TH1D*)gROOT->FindObject("hgb");
hg->Draw();
hgb->SetLineColor(kRed);
hgb->Draw("same");
c1->Draw();
```
